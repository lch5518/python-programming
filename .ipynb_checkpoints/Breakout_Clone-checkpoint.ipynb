```python
import SimpleGUICS2Pygame.simpleguics2pygame as simplegui 
import random
import math

WIDTH = 480
HEIGHT = 600

#block size
SIZE = (480,500)
FLOOR = 50


PAD_LOC = [100,10]
HALF_PAD_WIDTH = PAD_LOC[0]/2
paddle_pos = [WIDTH/2,0]

ball_pos = [paddle_pos[0], HEIGHT-(FLOOR+PAD_LOC[1]+20)]
BALL_VEL = 4
Ball_start = [0,0]

in_play = False
LIVES = 3

TOP_BORDER = 40

BLOCK_SIZE = (80,20)
BLOCK_INFO = (\
  {"color":"rgb(162, 71, 234)"	, "points":60},\
  {"color":"rgb(162, 71, 234)"	, "points":60},\
  {"color":"rgb(122, 157, 237)"	, "points":50},\
  {"color":"rgb(122, 157, 237)"	, "points":50},\
  {"color":"rgb(84, 237, 190)"	, "points":40},\
  {"color":"rgb(84, 237, 190)"	, "points":30},\
  {"color":"rgb(75, 219, 85)"	, "points":20},\
  {"color":"rgb(75, 219, 85)"	, "points":10},\
  )

LEVEL_SPEED_FACTOR = 1.1
LEVEL_MAX_SPEED_FACTOR = 2.0
POINT_MULTIPLIER = 1.1




# create info about a block
def block_info(pos, size, points, color):
    return {\
      "pos": tuple(pos),\
      "points": points,\
      "color": color,\
      "pos1": (pos[0]+1, pos[1]+size[1]/2),\
      "pos2": (pos[0]+size[0]-1, pos[1]+size[1]/2),\
      }


class Ball:
    def __init__(self,radius,ball_pos,ball_vel):
        self.radius = radius
        self.ball_pos = ball_pos
        self.ball_vel = ball_vel
        
    def draw(self,canvas,ball_pos):
        canvas.draw_circle(ball_pos, self.radius, 1,"White", "White")
        ball_angle = math.pi/2 + random.choice((-1,1)) * (math.pi*0.01 + math.pi*0.25 * random.randrange(100)/100)
        sv = BALL_VEL * min((LEVEL_SPEED_FACTOR**level), LEVEL_MAX_SPEED_FACTOR)
        ball_vel = [sv*math.cos(ball_angle), sv*math.sin(ball_angle)]

# prepare for a new game
def new_game():
    global lives, level, points
    
    lives = LIVES
    level = 0
    points = 0

    new_level()
        
# start a new level
def new_level():
    global level, blocks, paddle_pos, paddle_vel, ball_pos, ball_vel
    
    level += 1
    
    blocks = []
    x0 = (SIZE[0] % BLOCK_SIZE[0])/2
    ncols = SIZE[0]//BLOCK_SIZE[0]
    for row in range(len(BLOCK_INFO)):
        for col in range(ncols):
            blocks.append(block_info(\
              (x0+col*BLOCK_SIZE[0],TOP_BORDER+row*BLOCK_SIZE[1]),\
              BLOCK_SIZE,\
              level*BLOCK_INFO[row]["points"],\
              BLOCK_INFO[row]["color"]\
              ))
    paddle_vel = 0
    
def launch_ball():
    global Ball_start, ball_pos, ball_vel
    inplay = True
    Ball_start = [-3,-6]
    if lives==0:
        new_game()
    if len(blocks)==0:
        new_level()
        
def draw(canvas):
    global paddle_pos, ball_pos, ball_vel, blocks, Ball_start, points, point_multiplier, POINT_MULTIPLIER,PAD_LOC, block_to_remove
    
    #paddle move
    if paddle_pos[0] >= HALF_PAD_WIDTH:
        paddle_pos[0] += paddle_vel
    else:
        paddle_pos[0] = HALF_PAD_WIDTH
        
    if paddle_pos[0] <= WIDTH-HALF_PAD_WIDTH:
        paddle_pos[0] += paddle_vel
    else:
        paddle_pos[0] = WIDTH-HALF_PAD_WIDTH
    
    #draw Ball(반지름,공위치,공속도)
    ball = Ball(10,ball_pos,Ball_start)
    ball.draw(canvas,ball_pos)
    ball.ball_pos[0] += ball.ball_vel[0]
    ball.ball_pos[1] += ball.ball_vel[1]
    
    # check for ball hitting a block
    if ball.ball_pos[1]<=TOP_BORDER+len(BLOCK_INFO)*BLOCK_SIZE[1]: # optimization
        block_to_remove = None
        sx = ball.ball_pos[0]
        sy = ball.ball_pos[1]
        sl = ball.ball_pos[0]-ball.radius #left
        sr = ball.ball_pos[0]+ball.radius #right
        st = ball.ball_pos[1]-ball.radius #top
        sb = ball.ball_pos[1]+ball.radius #bottom
        svx = ball.ball_vel[0]
        svy = ball.ball_vel[1]
        
        for b in blocks:
            bl = b["pos"][0] #left
            br = b["pos"][0]+BLOCK_SIZE[0] #right
            bt = b["pos"][1] #top
            bb = b["pos"][1]+BLOCK_SIZE[1] #bottom
              
            if (bt<st<bb and bl<sx<br and svy<0) or (bt<sb<bb and bl<sx<br and svy>0):
                # bounce off bottom or top of block
                ball.ball_vel[1] *= -1
                block_to_remove = b
                break
            elif (bl<sr<br and bt<sy<bb and svx>0) or (bl<sl<br and bt<sy<bb and svx<0):
                # bounce off left or right of block
                ball.ball_vel[0] *= -1
                block_to_remove = b
                break
        point_multiplier=1
        if block_to_remove!=None:
            blocks.remove(block_to_remove)
            points += int(point_multiplier*block_to_remove["points"])
            point_multiplier *= POINT_MULTIPLIER
    
    # bounce off paddle 공<->패들
    if paddle_pos[0]<ball.ball_pos[0]<paddle_pos[0]+PAD_LOC[0] and paddle_pos[1]<ball.ball_pos[1]+ball.radius<paddle_pos[1]+PAD_LOC[1] and ball.ball_vel[1]>0:
        relpos = ball.ball_pos[0]-paddle_pos[0] /  PAD_LOC[0]
        if relpos<0.3 or relpos>0.7:
            ball.ball_vel[0] = -(ball.ball_vel[0]+0.5)
        ball.ball_vel[1] *= -1
        point_multiplier = 1
        print(point_multiplier)
    
    
    
    # bounce off paddle
    if (ball.ball_pos[1] <= ball.radius + PAD_LOC[1]) or (ball.ball_pos[1] >=  WIDTH-PAD_LOC[1]-ball.radius):    
        ball.ball_vel[1] = -ball.ball_vel[0]
    
    
    # bounce ball off left/right/top walls
    if (ball.ball_pos[0]-ball.radius<0 and ball.ball_vel[0]<0) or (ball.ball_pos[0]+ball.radius>SIZE[0] and ball.ball_vel[0]>0):
        ball.ball_vel[0] *= -1
    if ball.ball_pos[1]-ball.radius<0 and ball.ball_vel[1]<0:
        ball.ball_vel[1] *= -1
            
        
    
    
    #draw paddle 
    canvas.draw_line([paddle_pos[0]-HALF_PAD_WIDTH, HEIGHT-FLOOR], 
                     [paddle_pos[0]+HALF_PAD_WIDTH, HEIGHT-FLOOR], PAD_LOC[1], 'White')
    #paddle move
    if paddle_pos[0] >= HALF_PAD_WIDTH:
        paddle_pos[0] += paddle_vel
    else:
        paddle_pos[0] = HALF_PAD_WIDTH
        
    if paddle_pos[0] <= WIDTH-HALF_PAD_WIDTH:
        paddle_pos[0] += paddle_vel
    else:
        paddle_pos[0] = WIDTH-HALF_PAD_WIDTH
        
    #draw blocks
    for b in blocks:
        canvas.draw_line(b["pos1"], b["pos2"], BLOCK_SIZE[1]-2, b["color"])
    #print(b)
        
        
#keydown : 키보드를 누를 때
def keydown(key):
    global paddle_vel, ball_vel, ball_pos
    if key==simplegui.KEY_MAP["left"]:
        paddle_vel = -6
    if key==simplegui.KEY_MAP["right"]:
        paddle_vel = 6
    if key==simplegui.KEY_MAP["space"]:
        launch_ball()
        inplay = True
        
        
#keyup : 값을 입력중이 아닐 때 paddle속도는 0
def keyup(key):
    global paddle_vel
    if key==simplegui.KEY_MAP["left"]:
        paddle_vel = 0
    if key==simplegui.KEY_MAP["right"]:
        paddle_vel = 0
        
        
frame = simplegui.create_frame("Breakout clone", WIDTH, HEIGHT)
new_game()
frame.set_draw_handler(draw)
frame.set_keydown_handler(keydown)
frame.set_keyup_handler(keyup)
label = frame.add_label('Label')
label.set_text('Start space')

frame.start()
```
