버튼: higher, lower, correct<br>
범위[0,100)과 [0,1000)은  컴퓨터가 랜덤하게 선택<br>
나머지는 변함없음.<br>

컴퓨터는 항상 좁혀진 범위 내의 숫자를 선택하여야 한다.  
예를 들어 범위 [0,100)을 선택하고, 컴퓨터가  첫번째 guessing number 로 30을 선택한 경우, 플레이어가 lower 버튼을 누르면,
컴퓨터는 (0,29) 범위에서 숫자를 선택하여야 한다.  

[0,100)은 최대 7회, [0,1000)은 최대 10회 시도 할 수 있으며, 최대 시도 횟수를 넘게되면, 컴퓨터가 진다.<br><br><br>

**컴퓨터의 정수 선택 방법**  

(기본 값) 범위안에서 랜덤하게 정수 선택. 

한남이 버튼: 범위의 중간값에 가장 가까운 정수 선택<br> <br><br>

###  또는 새로운 응용 아이디어를 코딩

```python
import SimpleGUICS2Pygame.simpleguics2pygame as simplegui
import random


def new_game(guess):
    global gn
    print("Game start")
    n = int(guess)
    gn = random.randint(0,n)    
    

def range100():
    global gn, i, div, L, H
    H = 100
    L = 0
    i = 7
    print("0에서 100까지의 임의의 수")
    new_game(100)
    print("guessing number = ", gn)
    div = 100

def range1000():
    global gn, i, div, L, H
    H = 1000
    L = 0    
    i = 10
    print("0에서 1000까지의 임의의 수")
    new_game(1000)
    print("guessing number = ", gn)
    div = 1000



def output():
    global gn, i
    print("Guessing Number = {", gn,"}")
    print("남은 기회 : ",i)
    print("----------------")
    

def higher():
    global gn, H, L, i
    print("[higher]")
    
    L = gn + 1
    while(True):
        gn = random.randint(L,H)
        if(gn >= L and gn <= H):
            break
    print("min : ", L, "  max : ",H) 

    
    if(L == H):
        print("Correct!!!!! [",gn,"]")
        print("")
        if(div == 100):
            range100()
        elif(div ==1000):
            range1000()
    
    i -= 1
    if(i < 0):
        print("횟수 초과ㅠ")
        print("")
        print("restart")
        if(div == 100):
            range100()
        elif(div ==1000):
            range1000()
    
    output()

    
def lower():
    global gn, H, L, i
    print("[lower]")
    
    H = gn - 1
    while(True):
        gn = random.randint(L,H)
        if(gn >= L and gn <= H):
            break
    print("min : ", L, "  max : ",H) 

    if(L==H):
        print("Correct!!!!! [",gn,"]")
        print("")
        if(div == 100):
            range100()
        elif(div ==1000):
            range1000()
    
    i -= 1
    if(i < 1):
        print("횟수 초과ㅠ")
        print("")
        print("restart")
        if(div == 100):
            range100()
        elif(div ==1000):
            range1000()
    output()
            
def hannamH():
    global gn, H, L, i
    print("Hannam")
    
    L = gn + 1 
    print("min : ", L,"max: ", H)
    gn = round((H - L)/2) + L
    
    
    if(i < 1):
        print("횟수 초과ㅠ")
        print("")
        print("restart")
        if(div == 100):
            range100()
        elif(div ==1000):
            range1000()
    output()
    
def hannamL():
    global gn, H, L, i
    print("Hannam")
    
    H = gn  - 1
    print("min : ", L,"max: ", H)
    gn = round((H - L)/2) + L
    
    i -= 1
    if(i < 1):
        print("횟수 초과ㅠ")
        print("")
        print("restart")
        if(div == 100):
            range100()
        elif(div ==1000):
            range1000()
    output()
    
def correct():
    global gn
    print("Correct!!!!! [",gn,"]")
    print("")
    
    
    
f = simplegui.create_frame("up down game",300,300)

    
f.add_button("Range [0,100) Start", range100, 150)
f.add_button("Range [0,1000) Start", range1000, 150)

f.add_button("Higher", higher, 70)
f.add_button("Higher_Hanman", hannamH, 130)
f.add_button("Lower", lower, 70)
f.add_button("Lower_Hanman", hannamL, 130)
f.add_button("Correct", correct, 200)


f.start()
```
